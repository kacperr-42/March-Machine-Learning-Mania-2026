# Transformacja danych - March Madness 2026

**Output:** `m_data.csv` i `w_data.csv`

```python
import pandas as pd
from pathlib import Path

path = Path("data/")
```

```python
m_tourney = pd.read_csv("data/MNCAATourneyDetailedResults.csv")
w_tourney = pd.read_csv("data/WNCAATourneyDetailedResults.csv")
m_seeds = pd.read_csv("data/MNCAATourneySeeds.csv")
w_seeds = pd.read_csv("data/WNCAATourneySeeds.csv")
m_regular = pd.read_csv("data/MRegularSeasonDetailedResults.csv")
w_regular = pd.read_csv("data/WRegularSeasonDetailedResults.csv")
```

## Mężczyźni

```python
m_tourney["isTourney"] = 1
m_regular["isTourney"] = 0

m_tourney_regular = pd.concat([m_tourney, m_regular], ignore_index=True)

m_data = (
    m_tourney_regular
    .merge(m_seeds, left_on=["Season", "WTeamID"], right_on=["Season", "TeamID"], how="left")
    .rename(columns={"Seed": "WSeed"})
    .drop(columns=["TeamID"])
    .merge(m_seeds, left_on=["Season", "LTeamID"], right_on=["Season", "TeamID"], how="left")
    .rename(columns={"Seed": "LSeed"})
    .drop(columns=["TeamID"])
)

m_data.loc[m_data["isTourney"] == 0, ["WSeed", "LSeed"]] = pd.NA

print(f"m_data shape: {m_data.shape}")
m_data.head()
```

## Kobiety

```python
w_tourney["isTourney"] = 1
w_regular["isTourney"] = 0

w_tourney_regular = pd.concat([w_tourney, w_regular], ignore_index=True)

w_data = (
    w_tourney_regular
    .merge(w_seeds, left_on=["Season", "WTeamID"], right_on=["Season", "TeamID"], how="left")
    .rename(columns={"Seed": "WSeed"})
    .drop(columns=["TeamID"])
    .merge(w_seeds, left_on=["Season", "LTeamID"], right_on=["Season", "TeamID"], how="left")
    .rename(columns={"Seed": "LSeed"})
    .drop(columns=["TeamID"])
)

w_data.loc[w_data["isTourney"] == 0, ["WSeed", "LSeed"]] = pd.NA

print(f"w_data shape: {w_data.shape}")
w_data.head()
```

```python
m_data.to_csv("data/m_data.csv", index=False)
w_data.to_csv("data/w_data.csv", index=False)
```

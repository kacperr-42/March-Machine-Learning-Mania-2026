```python
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt

DIR_M = 'season_splits/M/strength_model/'
DIR_W = 'season_splits/M/strength_model/'
```

```python
m_train_str = pd.read_csv(DIR_M + 'train.csv')
```

```python
m_train_str.drop(['Gender', 'GameType', 'isRegularSeason', 'isTourney', 'WSeed', 'LSeed', 'WSeedNum', 'LSeedNum','WSeedRegion', 'LSeedRegion', '_row_order', '_is_tournament', ], axis=1)
```

```python


coach_counts = m_train_str['WCoachName'].value_counts()

plt.figure(figsize=(12,6))
coach_counts.head(20).plot(kind='bar')

plt.title('Top 20 trenerów (WCoachName) – liczność')
plt.xlabel('Coach Name')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
